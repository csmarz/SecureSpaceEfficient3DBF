### Bloom Filter Dua Dimensi

```python
#fungsi 2DBF

def AlokasiBloomFilter2(x,y):
  BF=[[0 for i in range(y)]for i in range(x)]
  return BF
def SetBit2(BF,i,j,pos):
  #mengeset bit dari 0 menjadi 1 sebagai tanda terisi
  d = BF[i][j]
  p = 1 << pos
  BF[i][j] = d|p
def TesBit2(BF,i,j,pos):
  #mengecek apakah ada isi atau tidak, bit 0 atau berisi bit 1 jika dalam biner. Jika ada = True, 0 = False
  d = BF[i][j]
  p = 1 << pos
  r = d ^ p
  t = r & p
  if t == 0 and d != 0:
    return True
  return False
def HitungIsiBF2():
  #menghitung isi bloom filter
  global hitung_input2
  hitung_input2 += 1
  return hitung_input2
def CekBF2(BF):
  #mengecek bloom filter penuh atau tidak
  if HitungIsiBF2() != maks2:
    return False
  else:
      return True
def CekAnggota2(BF,userpass):
  #proses mengecek anggota
  h = sha256(userpass.encode('utf-8'))
  h10 = int(h.hexdigest(),16)
  I = h10 % x
  J = h10 % y
  Pos = h10 % r
  cek = TesBit2(BF,I,J,Pos)
  if cek == True:
    return True
  return False 
def MasukdanCekAnggota2(BF,userpass):
  #proses memasukkan anggota
  h = sha256(userpass.encode('utf-8'))
  h10 = int(h.hexdigest(),16)
  I = h10 % x
  J = h10 % y
  Pos = h10 % r
  # print(I,J,Pos,h10)
  if CekBF2(BF) == False and CekAnggota2(BF,userpass) == False:
    HitungIsiBF2()
    SetBit2(BF,I,J,Pos)
    #pprint.pprint(BF)
  else:
    if CekBF2(BF) == True:
      print("2DBF Penuh")
    else:
      global fp2
      fp2 += 1
      # print("Sudah menjadi anggota")
```

### Bloom Filter Tiga Dimensi

```python
#fungsi 3DBF

def AlokasiBloomFilter(x,y,z):
  #mengalokasian tempat untuk penyimpanan dan penginisialisasian tempat dengan angka 0
  BF = [[[0 for i in range(z)]for i in range(y)]for i in range(x)]
  return BF
def SetBit(BF,i,j,k,pos):
  #mengeset bit dari 0 menjadi terisi (pos adalah desimal)
  d = BF[i][j][k] 
  p = 1 << pos
  BF[i][j][k] = d|p
  #print(d,p,BF[i][j][k])
def TesBit(BF,i,j,k,pos):
  #mengecek apakah ada isi atau tidak, bit 0 atau berisi bit 1 jika dalam biner. Jika ada = True, 0 = False
  d = BF[i][j][k]
  p = 1 << pos
  r = d ^ p
  t = r & p
  if t == 0 and d != 0:
    return True
  return False
def HitungIsiBF():
  #menghitung isi bloom filter
  global hitung_input
  hitung_input += 1
  return hitung_input
def CekBF(BF):
  #mengecek bloom filter penuh atau tidak
  if HitungIsiBF() != maks:
    return False
  else:
      return True
def CekAnggota(BF,userpass):
  #proses mengecek anggota
  h = sha256(userpass.encode('utf-8'))
  h10 = int(h.hexdigest(),16)
  I = h10 % x
  J = h10 % y
  K = h10 % z
  Pos = h10 % r
  cek = TesBit(BF,I,J,K,Pos)
  if cek == True:
    return True
  return False 
def MasukdanCekAnggota(BF,userpass):
  #proses memasukkan anggota
  h = sha256(userpass.encode('utf-8'))
  h10 = int(h.hexdigest(),16)
  I = h10 % x
  J = h10 % y
  K = h10 % z
  Pos = h10 % r
  if CekBF(BF) == False and CekAnggota(BF,userpass) == False:
    HitungIsiBF()
    SetBit(BF,I,J,K,Pos)
    # print("Menambah anggota")
  else:
    if CekBF(BF) == True:
      print("3DBF Penuh")
    else:
      global fp
      fp += 1
      # print("Sudah menjadi anggota")
```

## FPP
###### (tidak lanjut digunakan karena hanya dapat menghitunghasil angka yang kecil)

```python
import math 
import itertools
import pandas as pd
from google.colab import drive,files
import numpy as np

def product(a,b):
    # computing a * (a+1) * ... * (b-1) * b
    product = 1
    for i in range(a,b+1):
        product *= i
    return product

def perm(n,r): # r-permutation of n elements
    return product(n-r+1,n)

def comb(n,r): # r-combination of n elements
    return perm(n,r)//perm(r,r)

def power(a,n): 
    # computing a^n efficiently 
    # where a and n are integers, a is non-zero, n is non-negative
    if n == 0: return 1
    else:
        x = 1; power = a; quotient = n
        while quotient > 0:
            digit = quotient % 2
            if digit == 1: x = (x*power)
            power = power*power
            quotient = quotient //2
        return x

def fppinner(x,m,n): # inner loop of FPP calculation
    # sum j = 0 .. x of (-1)^j comb(x,j) ((x-j)/m)^n
    innersum = 0
    for j in range(0,x+1):
        if j % 2 == 0: innersum += comb(x,j) * (power((x-j)/m,n))
        else: innersum -= comb(x,j) * (power((x-j)/m,n))
    return innersum

def fpp(m,n): # computing fpp(m,n) using the formula
    # sum x = 0 .. m of (x/m)*comb(m,x)*fppinner(x,m,n)
    fppsum = 0
    for x in range(0,m+1):
        fppsum += (x/m) * comb(m,x) * fppinner(x,m,n)
    return fppsum
```
